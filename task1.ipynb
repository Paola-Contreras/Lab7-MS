```python
# Imports 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Generaci√≥n de datos  
n = 100
presupuesto = np.linspace(100, 1000, n)

# Generar datos de CTR relacionados con el presupuesto 
CTR = 2.5 + 0.025 * presupuesto 

# Agregar ruido para simular variabilidad
ruido = np.random.normal(0, 2, n)
CTR += ruido

#------------------------------------------------------------------------------------------------------------
# Crear un modelo sustituto utilizando una regresi√≥n polin√≥mica
degree = 2  # Grado del polinomio
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(presupuesto.reshape(-1, 1))
model = LinearRegression()
model.fit(X_poly, CTR)

#------------------------------------------------------------------------------------------------------------
# Realizar predicciones
presupuesto_pred = np.linspace(100, 1000, 100)  # Valores de presupuesto para predicci√≥n
X_pred = poly_features.transform(presupuesto_pred.reshape(-1, 1))
CTR_pred = model.predict(X_pred)

#------------------------------------------------------------------------------------------------------------
# Graficar  
plt.figure(figsize=(10, 6))
plt.scatter(presupuesto, CTR, label="Datos reales", color="#57abf0")
plt.plot(presupuesto_pred, CTR_pred, label="Predicciones", color="#ee5962")
plt.xlabel("Presupuesto")
plt.ylabel("Tasa de Clics (CTR)")
plt.title("Modelo Sustituto de Regresi√≥n Polin√≥mica")
plt.legend()
plt.grid(True)
plt.show()

```

- ¬øPor qu√© este es un ejemplo de meta-modelado?
    - Este es un ejemplo de meta modelo puesto a que se emplea una regresi√≥n pol√©mica para modelar la relaci√≥n entre la tasa de clics y el presupuesto. En este caso se hace uso de un polinomio de segundo grado para representar dicha relaci√≥n
